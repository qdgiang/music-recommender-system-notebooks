```python
import pandas as pd
```

```python
def get_pivot_data():
    triplets = pd.read_csv('data/triplets_file.csv')
    songs = triplets.groupby('song_id').sum().sort_values('listen_count', ascending=False)
    songs = songs[songs['listen_count'] < 100]

    # delete songs with less than 100 listens
    triplets = triplets[~triplets['song_id'].isin(songs.index)]

    # all listen counts greater than 130 are set to 130
    triplets['listen_count'] = triplets['listen_count'].apply(lambda x: 130 if x > 130 else x)
    
    # use the first 500000 rows
    triplets = triplets.head(1000000)

    # add personal data
    me = pd.read_csv('data/me.csv')
    triplets = pd.concat([me, triplets], ignore_index=True)
    
    # get list of songs and total listen count
    songs = triplets.groupby('song_id').sum().sort_values('listen_count', ascending=False)
    return triplets.pivot(index='song_id', columns='user_id', values='listen_count').fillna(0), songs
```

```python
song_data = pd.read_csv('data/meta_data.csv')
```

```python
pivot, songs_list = get_pivot_data()
```

```python
import numpy as np
```

```python
pivot_norm = pivot.apply(lambda x: x - np.nanmean(x), axis=1)
```

```python
pivot_norm
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
item_sim_df = pd.DataFrame(cosine_similarity(pivot_norm, pivot_norm), index=pivot_norm.index, columns=pivot_norm.index)
item_sim_df.head()
```

```python
# export item similarity matrix to pickle file
item_sim_df.to_pickle('model/item_sim_df.pkl')
```

```python
def get_song_id(title, artist_name):
    return song_data[song_data.artist_name == artist_name][song_data.title == title]['song_id'].values[0]
```

```python
def get_similar_song(song_id):
    if song_id not in pivot_norm.index:
        print(None)
        return None, None
    else:
        sim_songs = item_sim_df.sort_values(by=song_id, ascending=False).index[1:]
        sim_score = item_sim_df.sort_values(by=song_id, ascending=False).loc[:, song_id].tolist()[1:]
        return sim_songs, sim_score
```

```python
ID = get_song_id('The Scientist', 'Coldplay')
#print(type(ID))
#print(ID)
songs, scores = get_similar_song(ID)

# map x to song title
for x,y in zip(songs[:10], scores[:10]):
    print("{}, by {} with similarity of {}".format(
        song_data[song_data.song_id == x].title.values[0],
        song_data[song_data.song_id == x].artist_name.values[0],
        y))
```
