```python
import pandas as pd
```

```python
triplets = pd.read_csv('data/triplets_file.csv')
me = pd.read_csv('data/me.csv')
```

```python
# add me to the triplets
triplets = triplets.append(me, ignore_index=True)
```

```python
# export triplets to pickle
triplets.to_pickle('model/triplets.pkl')
```

```python
USER_ID = '1234_thisisme'
```

```python
# get user history
user_history = triplets[triplets['user_id'] == USER_ID]
user_history
```

```python
popularity_based = triplets.groupby('song_id')['listen_count'].sum().reset_index()
popularity_based = popularity_based.sort_values('listen_count', ascending=False)
popularity_based
```

```python
# recommend top 10 songs that the user hasn't listened to
recommendations = popularity_based[~popularity_based['song_id'].isin(user_history['song_id'])]
recommendations.head(10)
```

```python
# add song names to recommendations
metadata = pd.read_csv('data/song_data.csv')
recommendations = pd.merge(recommendations, metadata.drop_duplicates(['song_id']), on='song_id', how='left')
#recommendations = recommendations.drop(['song_id', 'listen_count'], axis=1)
recommendations.head(10)
```

```python
recommendation_3_artists = recommendations[recommendations['artist_name'].isin(['Maroon 5', 'Coldplay', 'Justin Bieber'])].reset_index()
recommendation_3_artists.drop(['index', 'song_id', 'listen_count'], axis=1, inplace=True)
recommendation_3_artists.head(20)

```
