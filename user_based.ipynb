```python
def get_pivot_data():
    triplets = pd.read_csv('data/triplets_file.csv')
    songs = triplets.groupby('song_id').sum().sort_values('listen_count', ascending=False)
    songs = songs[songs['listen_count'] < 100]

    # delete songs with less than 100 listens
    triplets = triplets[~triplets['song_id'].isin(songs.index)]

    # all listen counts greater than 130 are set to 130
    triplets['listen_count'] = triplets['listen_count'].apply(lambda x: 10 if x > 10 else x)
    
    # use the first 500000 rows
    triplets = triplets.head(500000)

    # add personal data
    me = pd.read_csv('data/me.csv')
    triplets = pd.concat([me, triplets], ignore_index=True)
    
    # get list of songs and total listen count
    songs = triplets.groupby('song_id').sum().sort_values('listen_count', ascending=False)
    return triplets.pivot(index='user_id', columns='song_id', values='listen_count').fillna(0), songs
    #return triplets.pivot(index='user_id', columns='song_id', values='listen_count'), songs
```

```python
def get_top10_recommended_songs(user_id):
    # get the user's row, cut off by the threshold
    user_row = user_sim_df.loc[user_id]
    #user_row = user_row[user_row > thresh_hold]
    user_row.drop(index=user_id, inplace=True)
    # get the top n users
    user_row = user_row.sort_values(ascending=False).head(20)
    # get all songs ID that the top n users have listened to. Drop all 0s. Drop all duplicates. Return as a list
    songs = pivot.loc[user_row.index]
    #print(songs.shape)
    songs = songs[songs > 0]
    songs = songs.dropna(axis=1, how='all')
    songs = songs.columns.drop_duplicates().tolist()

    target_user = pivot.loc[user_id]
    target_user = target_user[target_user > 0]
    #songs = [song for song in songs if song not in target_user.index]
    #print(len(songs))

    # get the pivot table of the songs and the top n users
    pivot_songs = pivot.loc[user_row.index, songs]
    pivot_songs.head()
    pivot_songs.shape

    # Weighted average of the top n users' listen count, using the similarity as the weight. Sort descending
    weighted_avg = pivot_songs.mul(user_row.values, axis=0).sum(axis=0) / user_row.sum()
    weighted_avg = weighted_avg.sort_values(ascending=False)
    weighted_avg.head(10)

    # add song title and artist name
    #weighted_avg = weighted_avg.to_frame()
    #weighted_avg = weighted_avg.merge(song_data, left_index=True, right_on='song_id')
    #weighted_avg = weighted_avg[['song_id', 'title', 'artist_name']]
    return weighted_avg.head(10)
```

```python
# for each user, recommend the top 10 songs. Compare the top 10 songs with the test data
# and calculate the precision and recall
print(test.shape)
def get_precision_recall(test, user_sim_df, top_n=10):
    precision = []
    recall = []
    for user in test.index:
        #print(user)
        # get the top 10 recommended songs
        top10 = get_top10_recommended_songs(user)
        #print(top10.index)
        # top 10 contains 2 columns: song_id and predicted listen count
        # get the songs that the user has listened to
        target_user = test.loc[user]
        target_user = target_user[target_user > 0]
        # calculate the precision and recall
        p = len(set(top10.index).intersection(set(target_user.index))) / top_n
        print(p)
        r = len(set(top10.index).intersection(set(target_user.index))) / len(target_user)
        print(r)
        precision.append(p)
        recall.append(r)

    return np.mean(precision), np.mean(recall)
```

```python
# calculate map@k
def get_mapk(test, user_sim_df, k=10):
    mapk = []
    for user in test.index:
        top10 = get_top10_recommended_songs(user)
        target_user = test.loc[user]
        target_user = target_user[target_user > 0]
        p = len(set(top10.index).intersection(set(target_user.index))) / k
        r = len(set(top10.index).intersection(set(target_user.index))) / len(target_user)
        mapk.append(p * r)
    return np.mean(mapk)
```

```python
get_mapk(test, user_sim_df)
```

```python
def get_mae(test, user_sim_df):
    mae = []
    for user in test.index:
        top10 = get_top10_recommended_songs(user)
        target_user = test.loc[user]
        target_user = target_user[target_user > 0]
        mae.append(np.mean(np.abs(top10 - target_user)))
    return np.mean(mae)
```

```python
get_mae(test, user_sim_df)
```

```python
def get_rmse(test, user_sim_df):
    rmse = []
    for user in test.index:
        top10 = get_top10_recommended_songs(user)
        target_user = test.loc[user]
        target_user = target_user[target_user > 0]
        rmse.append(np.sqrt(np.mean(np.square(top10 - target_user))))
    return np.mean(rmse)
```

```python
# split train and test set according to each user's listening history
icount = 200
def split_train_test(pivot):
    train = pd.DataFrame(columns=pivot.columns)
    test = pd.DataFrame(columns=pivot.columns)
    for user in pivot.index:
        global icount
        icount -= 1
        if icount < 0:
            break
        print(icount)
        user_row = pivot.loc[user]
        user_row = user_row[user_row > 0]
        if len(user_row) > 1:
            train_row, test_row = train_test_split(user_row, test_size=0.2, random_state=42)
            train = pd.concat([train, train_row.to_frame().T])
            test = pd.concat([test, test_row.to_frame().T])
    return train, test
```
